```python
import pandas as pd
data={
'x':[23,43,53,65,32,43,44,90],
'y':[78,32,54,67,87,89,97,99]
}
data=pd.DataFrame(data)
print(data)
matrix=data.corr()
print(matrix)
```

```python
import numpy as np
x=[23,43,53,65,32,43,44,90]
y=[78,32,54,67,87,89,97,99]

matix=np.corrcoef(x,y)
print(matix)
```
